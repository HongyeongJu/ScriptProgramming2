## [3번 문제]
다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

- 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.

 - count: Counter 인스턴스가 지니는 초기 정수 값
 - step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)

- 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.

 - cincr() 
 - d.incr() 
 - print(c) 
 - [Count (step: 1)] 11 
 - print(d) 
 - [Count (step: 2)] 12 
- 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.

 - c() 
 - d() 
 - print(c) 
 - [Count (step: 1)] 12 
 - print(d) 
 - [Count (step: 2)] 14 
- 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

 - c = c + 5 
 - d = d - 5 
 - print(c) 
 - [Count (step: 1)] 17 
 - print(d) 
 - [Count (step: 2)] 9 
- 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

 - print(c) > 10 
 - True 
 - print(d) > 10 
 - False 
 - print(c) < 10 
 - False 
 - print(d) < 10 
 - True 
 - print(c) == 17 
 - True  
 - print(d) != 9 
 - False 

```python
class Counter:
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
    def __str__(self):
        return "[Count(step: %d)] %d" % (self.step, self.count)
    def incr(self):
        self.count += self.step
    def __call__(self):
        self.incr()
    def __add__(self, other):
        return self.count + other
    def __sub__(self, other):
        return self.count - other
    def __lt__(self, other):
        return self.count < other
    def __eq__(self, other):
        return self.count == other
    def __gt__(self, other):
        return self.count > other
```

```python
c = Counter(10)
d = Counter(10, 2)
print(c)
print(d)
print()

c.incr()
d.incr()
print(c)
print(d)
print()

c()
d()
print(c)
print(d)
print()

c= c+5
d = d-5
print(c)
print(d)
print()

print(c > 10)
print(d > 10)
print(c < 10)
print(d < 10)
print(c == 17)
print(d != 9)
```

## [4번 문제]
다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

```python
class MySet(list):        # 클래스 정의 (list를 상속받는다.)
    def __init__(self, l): # 인스턴스 객체 리스트를 인자로 받는다.
        for e in l:        # append를 사용해서(list를 상속받아서 가능) 인자를 다 입력 받는다.
            self.append(e)
        MySet.eliminate_duplicate(self) # 중복되는 것 삭제하는 함수 호출 정적 메소드이므로 호출시엔 클래스 명. 붙여야됨
    def __str__(self):  # 객체를 print로 호출하면 출력하는 함수이다.
        result = "MySet: {"
        for e in self:
            result = result +str(e) + " ,"
        result = result[0:len(result)-2] + "}"  # 마지막  , <-을 삭제 하기 위해 -2를 해준것
        return result
    
    @staticmethod   # 정적메소드 호출하기 위해선  맨 앞에 클래스명.을 붙여야된다.
    def eliminate_duplicate(l):
        s = []   # 공백 리스트 만든다.
        for e in l:   # 인자로 받은 리스트의 인자를 호출해서 만약 리스트 안에 없다면 추가해서 중복을 검사
            if e not in s:
                s.append(e)
        l[:] =[]    # 인자로 받은 l 리스트를 공백 리스트로 만들고 거기에 대입해준다.
        for e in s:
            l.append(e)
                
if __name__ == "__main__": # 지금 현재 이름공간이 main함수이면 실행
    s = MySet([1,2,2,3])
    print(s)
    t = MySet([2,3,4,5,6,7,8,8,8,8,8,9])
    print(t)
```

## 코드 설명
- init은 생성자이다 따라서 객체를 만들었을 때 자동으로 호출되는 메소드이다. 위 코드는 호출시 eliminate_duplicate를 호출해서 중복되는 것을 검사한다.
- str은 print같은 함수를 이용해서 객체를 매개변수로 받았을 때 호출되는 함수이다. 이 함수는 자신의 상태를 문자열로 출력 할 수가 있는 점에서 간편하다.
- eliminate_duplicate메소드는 중복체크하는 메소드이다. 먼저 빈 리스트를 만들고 그 리스트 안에 인자가 있는지 없는지 확인 후 넣는 메소드이다.

## [5번 문제]
- 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

- 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
- 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

 - u = s | t 
 - print(u) 
 - MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} 
- 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

 - u = s & t 
 - print(u) 
 - MySet: {2, 3} 
- 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

 - s = MySet([1, 2, 3]) 
 - t = MySet([3, 4, 5]) 
 - u = s - t 
 - print(u) 
 - MySet: {1, 2} 

```python
class MySet(list):        # 클래스 정의 (list를 상속받는다.)
    def __init__(self, l): # 인스턴스 객체 리스트를 인자로 받는다.
        for e in l:        # append를 사용해서(list를 상속받아서 가능) 인자를 다 입력 받는다.
            self.append(e)
        MySet.eliminate_duplicate(self) # 중복되는 것 삭제하는 함수 호출 정적 메소드이므로 호출시엔 클래스 명. 붙여야됨
    def __str__(self):  # 객체를 print로 호출하면 출력하는 함수이다.
        result = "MySet: {"
        for e in self:
            result = result +str(e) + " ,"
        result = result[0:len(result)-2] + "}"  # 마지막  , <-을 삭제 하기 위해 -2를 해준것
        return result
    
    @staticmethod   # 정적메소드 호출하기 위해선  맨 앞에 클래스명.을 붙여야된다.
    def eliminate_duplicate(l):
        s = []   # 공백 리스트 만든다.
        for e in l:   # 인자로 받은 리스트의 인자를 호출해서 만약 리스트 안에 없다면 추가해서 중복을 검사
            if e not in s:
                s.append(e)
        l[:] =[]    # 인자로 받은 l 리스트를 공백 리스트로 만들고 거기에 대입해준다.
        for e in s:
            l.append(e)
            
    def __or__(self, other):
        l = MySet([])   # 공백 리스트를 인자로 받는 myset 설정
        for e in self:  # 먼저 자신의 리스트의 원소들을 다 넣고
            l.append(e)
        for e in other:  # 리스트에 없을 때 넣는다.
            if e not in l:
                l.append(e)
                
        return l
    
    def __sub__(self, other):  
        l = MySet([]) #공백 리스트를 인자로 받는 myset 설정
        for e in self:   # 먼저 왼쪽 피연산자의 리스트의 원소를 다 넣고
            l.append(e)
        for e in other:  # 만일 있을 때 왼쪽 피연산자의 원소를 삭제한다.
            if e in l:
                l.remove(e)
        return l
    
    def __and__(self, other):
        l = MySet([]) #공백 리스트를 인자로 받는 myset 설정
        for e in self:   # 왼쪽 피연산자와 오른쪽 피연산자의 원소들이 둘다 있을 때 원소를 넣는다.
            if e in other:
                l.append(e)
        return l
                
if __name__ == "__main__": # 지금 현재 이름공간이 main함수이면 실행
    s = MySet([1,2,2,3])
    print(s)
    t = MySet([2,3,4,5,6,7,8,8,8,8,8,9])
    print(t)
    print()
    
    u = s | t
    print(u)
    print()
    
    u = s & t
    print(u)
    print()
    
    s= MySet([1,2,3])
    t= MySet([3,4,5])
    u = s - t
    print(u)
    print()
```

먼저 각 메소드를 해결하기 위해서는 list가 필요한데 그냥 list를 사용하면 print(u) 를 사용해서 자신이 정의한 문자열을 출력을 할 수가 없으므로 list가 아니라 MySet([]) <- 공백 리스트를 인자로 받는 자신의 클래스를 호출한다. 그리고 이 빈 MySet리스트로 각각의 메소드데로 논리적으로 구현한다

```python
from urllib import request
import string
class WebWordsFrequency:
    def __init__(self, url = 0, *urllist):
        self.d = {} # 이 사전은 후에 url을 입력 받을 시 url에 해당되는 사전을 출력하는 것
        self.url1= []
        if url == 0 :
            return
        self.url1.append(url)
        for s in urllist:
            self.url1.append(s)
    def addUrl(self, url):
        self.url1.append(url)
    def removeUrl(self, url):
        if url in self.url1:
            self.url1.remove(url)
    def listUrls(self):
        for s in self.url1:
            print(s)
    def getWordsFrequency(self):
        for s in self.url1:
            dd = {};   # 현재 유알엘에 있는 용어들 저장 및 카운트 저장
            #print(str(s))
            f = request.urlopen(s)
            source_bytes = f.read()
            source = source_bytes.decode('utf-8')
            #print(source)

            # print(len(source))
            flag = False
            cnt = 0
            str1 = ' '
            for x in source :
                if (x == '<') or (flag == True) : 
                    flag = True
                    if x == '<' :
                        cnt = cnt + 1
                    if x == '>'  :
                        cnt = cnt - 1
                    #flag = False
                    if cnt == 0 :
                        flag = False
                    continue
                elif flag == False :
                    str1 = str1 + x

            while True : 
                str1 = str1[:str1.find('-->')] + str1[str1.find('-->') + 3:]
                if str1.find('-->') == -1 :
                    break;
#print(string.punctuation)
            str2 =''
            for x in str1 :
                if(x not in list(string.punctuation)) :
                    str2 = str2 + x
                else :
                    str2 = str2 + " "
        
# 사전 정의
            for x in str2.split() :
                dd[x] = str2.split().count(x)
            
            #불용어 리스트 추려내기!!
            li = []
            
            with open("nouse.txt", "r") as f:
                for i in f:
                    if i != '' :
                        li.append(i.strip())
            for i in range(li.count("")):
                li.remove("")
                
            # 사전 객체에 불용어 리스트를 삭제하기
            li1 = li.copy()
            
            for i in li :
                li1.append(i.upper())
                li1.append(i.capitalize())
                
            for i in li1:
                if i in dd:
                    del dd[i]
            #print(dd)
            # 사전 객체에 있는 사전에 주소와 사전 맵핑시키기
            
            self.d[s] = dd
        print(self.d.values())
            
    def getMaxFrequencyWords(self):
        max = 0
        print(self.url1)
        for u in self.url1:
            t = self.d[u].values()
            for a in t :  #max값 보다 큰지 확인
                if max < a :
                    max = a
        # max 값의 단어를 찾아야된다.
        # 하기전에 키값과 값이 바뀌어야된다.
        for u in self.url1:
            dic1 = self.d[u]
            dic2 = dict(zip(dic1.values(), dic1.keys())) # 사전 서로 바꾸기
            if max in dic2:
                print(dic2[max])
        if max == 0:
            return None
    def searchUrlByWord(self, instr):
        dic1 ={}
        for i in self.url1:
            if instr in self.d[i]:
                dic1[i] = self.d[i][instr]
        
        dic2 = dict(zip(dic1.values(), dic1.keys())) # 사전 정렬을 위해 값과 키값이 바뀜
        
        li1 = list(dic2.keys())
        li1.sort(reverse = True) #정렬한다
        j = 1
        for i in li1:
            print("우선도 : ", end = "")
            print(j, end = " ")
            print(dic2[i])
            j = j +1
```

## [8번 문제]
- 위 7번 문제에서 정의한 WebWordsFrequency 클래스를 상속하여 OrderedWebWordsFrequency 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드 하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.

- 리스트 내의 각 원소는 단어와 빈도를 쌍으로 지니는 튜플이다.
getWordFrequence() 메소드에 reverse 라는 인자를 만들고 true 또는 false를 인자로 받을 수 있도록 한다.

- reverse 인자의 디폴트 값은 false 이며, 기본적으로 내림차순으로 정렬한다.
- reverse 인자에 true를 넣으면 오름차순으로 정렬한다.
- w4 = OrderedWebWordsFrequency('http://www.times.com', 'https://www.amazon.com', 'https://github.com')

- w4.getWordsFrequency()
[('site', 12), ('science', 11), ('hello', 8), ('world', 2), ('program', 1), ('python', 1)]

- w4.getWordsFrequency(reverse=true)
[('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)]

```python
class OrderedWebWordsFrequency(WebWordsFrequency):
    def getWordsFrequency(self, reverse = False):
        for s in self.url1:
            dd = {};   # 현재 유알엘에 있는 용어들 저장 및 카운트 저장
            #print(str(s))
            f = request.urlopen(s)
            source_bytes = f.read()
            source = source_bytes.decode('utf-8')
            #print(source)

            # print(len(source))
            flag = False
            cnt = 0
            str1 = ' '
            for x in source :
                if (x == '<') or (flag == True) : 
                    flag = True
                    if x == '<' :
                        cnt = cnt + 1
                    if x == '>'  :
                        cnt = cnt - 1
                    #flag = False
                    if cnt == 0 :
                        flag = False
                    continue
                elif flag == False :
                    str1 = str1 + x

            while True : 
                str1 = str1[:str1.find('-->')] + str1[str1.find('-->') + 3:]
                if str1.find('-->') == -1 :
                    break;
#print(string.punctuation)
            str2 =''
            for x in str1 :
                if(x not in list(string.punctuation)) :
                    str2 = str2 + x
                else :
                    str2 = str2 + " "
        
# 사전 정의
            for x in str2.split() :
                dd[x] = str2.split().count(x)
            
            #불용어 리스트 추려내기!!
            li = []
            
            with open("nouse.txt", "r") as f:
                for i in f:
                    if i != '' :
                        li.append(i.strip())
            for i in range(li.count("")):
                li.remove("")
                
            # 사전 객체에 불용어 리스트를 삭제하기
            li1 = li.copy()
            
            for i in li :
                li1.append(i.upper())
                li1.append(i.capitalize())
                
            for i in li1:
                if i in dd:
                    del dd[i]
            #print(dd)
            # 사전 객체에 있는 사전에 주소와 사전 맵핑시키기
            
            self.d[s] = dd
        
        
        def mycmp(a):   # 숫자로 정렬하기 위한 비교 함수
            return a[1]
        
        # 종합 사전 객체 d , 유알엘 리스트 url1
        freli = [] # 튜플을 저장할 리스트 
        for u in self.url1:
            for i in self.d[u]:
                t = (i, self.d[u][i]) #튜플 형식으로 저장하기 위해 임시 튜플 생성
                freli.append(t)  # 튜플을 추가한다.
        if reverse == True: #만일 reverse 값이 True면 
            freli.sort(key = mycmp) # 오름 차순정렬
        elif reverse == False:  #만일 reverse 값이 false나 기본값이면
            freli.sort(key = mycmp, reverse = True) #내림차순 정렬
            
        print(freli)
    def __iter__(self):
        self.index = 0  # 리스트 참조를 하기위한 index값
        self.dl = []    # 새로운 공백 리스트를 만든다.
        for dd  in self.d.values():  ## 사전들 dd로 받는다.
            for aa in dd:   # 사전들의 키값들을 인자로 받는다.
                self.dl.append((aa,dd[aa]))  #키값과 사전의 값을 튜플로 묶어서 리스트에 넣는다.
        return self  # 반복자 참조 변수를 반환
    def __next__(self):
        if self.index >= len(self.dl): # 만일 index 값이 리스트의 길이를 넘으면 StopIteration 을 발생시킨다.
            raise StopIteration  # StopIteration이 발생되면 for문에서 중단된다.
        
        n = self.dl[self.index]  # 각각의 리스트의 요소를 n에 넣고
        self.index += 1   # 리스트를 증가시킨뒤
        return n  # n값을 반환한다.
    
```
