# 문제 4

다음 문자열을 ':'을 기준으로 분리하여 리스트로 만들고 각 문자열의 좌우 공백을 제거하시오 (즉, 문자열 S에서 l을 만들어라)
- s = '  first star   :   second star   :    third star  '
- l = ['first star', 'second star', 'third star']
    - 1) for ~ in 문을 이용한 코드를 작성하시오.
    - 2) 리스트 내포(list comprehension)를 이용한 코드를 작성하시오.

```python
s = 'first star : second star : third star'
l =[]
s1 = ''
for x in s :
    s1 = s1 +x
    if x == ':' :
        s1 =s1[:len(s1)-1]
        l.append(s1)
        s1 = ''
    
        
l.append(s1)
l2 = []
for x in l :
    x = x.strip()
    l2.append(x)
print(l2)
```

## 코드 설명
for 문을 사용해서 s의 값을 추출해서 새로운 문자열에 더하고 만약에 : 가 있을 경우 새로운 문자열에서 슬라이싱을 이용해서 마지막 꽁무니를 제거하여 : 를 삭제시켜준다. 그리고 :를 기준점으로 삼았기 때문에 l.append(s1)를 for문이 끝나고 추가를해주고 마지막으로 strip함수를 통해서 공백을 제거한다.

```python
s = 'first star : second star : third star'
l1 = []
l = [l1.append(x.strip()) for x in s.split(":")]
print(l1)
```

## 코드 설명
문자열 s의 문자열을 :기준점으로 나눈 리스트를 가지고 l1(공백 문자열)에 더하는데 이때 strip()함수를 사용해서 공백이 없는 리스트를 문자열에 더해준다.

[참고 1]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)

[참고 2]: sum() 등의 내장함수 사용 가능함

### 1

```python
#def addall(l):
#    sum = 0
#    for x in l :
#        sum = sum + x
#    return sum

def addall(l):
    L = [x for x in l ]
    return sum(L)
    
print(addall([1]))
print(addall([1,2,3,4,5,6,7,8,9]))
```

## 2

```python
#def addallodd(l) :
#    sum = 0
#    for x in l :
#        if (x % 2) == 1:
#            sum = sum + x
#    return sum


def addallodd(l) :
    L = [x for x in l if x % 2 == 1]
    return sum(L)
print(addallodd([1]))
print(addallodd([1,2,3,4,5,6,7,8,9]))
```

## 코드 설명
for문을 사용한 함수는 sum 변수를 만들어서 각각의 리스트의 원소의 합을 더해 결과를 출력하고 리스트 내포를 사용한 함수는 리스트 내포를 이용해서 각각의 리스트를 만든 뒤 sum 함수를 사용해서 그 결과 값을 리턴하도록한다. 리스트르 새롭게 만들 때 리스트 내포를 사용하면 쉽고 간편하게 만들 수 있음을 알 수가 있다.|

사전 d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}가 주어졌을 때 다음 요구사항에 맞는 코드를 제시하시오
- [참고]: d.keys(), d.values()를 통해 리스트를 얻어낸 후 리스트가 지니고 있는 sort(cmp=None, key=None, reverse=False)함수를 활용하시오.


- 1) 키의 알파벳 오름차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.
- 2) 키의 알파벳 내림차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.
- 3) 값의 오름차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.
- 4) 값의 내림차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.

### 1번

```python
d = {'one':1, 'two':2 , 'three':3, 'four':4 , 'five':5}
l = list(d.keys())
l.sort()
print(l)
for x in l:
    print((x, d[x]))
```

### 2번

```python
d = {'one':1, 'two':2 , 'three':3, 'four':4 , 'five':5}
l = list(d.keys())
l.sort(reverse = True)
print(l)
for x in l:
    print((x, d[x]))
```

## 1번과 2번 코드 설명
1번과 2번 코드는 전반적으로 거의 비슷하다. 먼저 새로운 리스트를 만드는데 이것은 사전의 키값으로 만들어낸다. 그리고 그 리스트를 정렬하고 이것을 토대로 for문을 사용해서 튜플형식으로 출력해주면된다. 2번 코드에서는 내림차순으로 출력해야되므로 sort함수에 reverse = True를 추가해줘서 내림차순을 구현한다

### 3번

```python
d = {'one':1, 'two':2 , 'three':3, 'four':4 , 'five':5}
d2 = dict(zip(d.values(), d.keys()))

l = list(d2.keys())
l.sort()
print(l)
for x in l:
    print((x, d2[x]))
```

### 4번

```python
d = {'one':1, 'two':2 , 'three':3, 'four':4 , 'five':5}
d2 = dict(zip(d.values(), d.keys()))

l = list(d2.keys())
l.sort(reverse =True)
print(l)
for x in l:
    print((x, d2[x]))
```

## 3번과 4번 코드 설명
3번과 4번 코드는 1번 2번 코드와 한가지만 빼고 같다. 그 한가지는 d2를 만들어서 키값과 벨류값을 서로 바꿔주는 것인데, 서로 바꾸기 위해서 zip함수를 사용하여 zip(d.values(), d.keys()) 같이 키값 리스트와 벨류값 리스트를 반대로 입력시켜 새로운 사전을 만들고 그 사전을 바탕으로 숫자를 기준으로해서 오름차순과 내림차순 정렬을 실시한다.

'http://cse.koreatech.ac.kr' 사이트

```python
from urllib import request
import string
url = 'http://cse.koreatech.ac.kr'
f = request.urlopen(url)
source_bytes = f.read()
source = source_bytes.decode('utf-8')
# print(len(source))
flag = False
cnt = 0
str1 = ' '
for x in source :
    if (x == '<') or (flag == True) : 
        flag = True
        if x == '<' :
            cnt = cnt + 1
        if x == '>'  :
            cnt = cnt - 1
            flag = False
        if cnt == 0 :
            flag = False
        continue
    else :
        str1 = str1 + x

while True : 
    str1 = str1[:str1.find('-->')] + str1[str1.find('-->') + 3:]
    if str1.find('-->') == -1 :
        break;
#print(string.punctuation)
str2 =''
for x in str1 :
    if(x not in list(string.punctuation)) :
        str2 = str2 + x
    else :
        str2 = str2 + " "
        
# 사전 정의
d = {}
for x in str2.split() :
    d[x] = str2.split().count(x)
    
print(d)
#print(str2.split())
#print(len(str1.split()))
```
